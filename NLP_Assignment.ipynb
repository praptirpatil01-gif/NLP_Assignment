Question 1: What is Computational Linguistics and how does it relate to NLP?
Answer:

Computational Linguistics (CL) is an interdisciplinary field that combines linguistics, computer science, and artificial intelligence to study and model human language computationally. It focuses on creating algorithms and systems that can understand, analyze, and generate natural language.

Relationship with NLP:

Computational Linguistics provides the theoretical foundation for NLP.
NLP uses computational linguistic principles to build practical applications.
CL focuses on language structure and theory, while NLP emphasizes implementation and real-world solutions.

Example:
Grammar rules developed in Computational Linguistics are used by NLP systems for machine translation and text analysis.

Conclusion:
Computational Linguistics and NLP are closely related fields. Computational Linguistics provides linguistic theories, while NLP applies those theories to develop intelligent language-processing systems.

Question 2: Briefly describe the historical evolution of Natural Language Processing.
Answer:

The evolution of NLP can be divided into several phases:

1. Rule-Based Era (1950s–1980s)
NLP systems relied on manually created grammar and linguistic rules.
Early machine translation systems were developed.
2. Statistical NLP Era (1980s–2000s)
Machine learning techniques began replacing hand-crafted rules.
Large text corpora were used for language modeling.
3. Machine Learning Era (2000s–2015)
Algorithms such as Naive Bayes, SVM, and Decision Trees became popular.
Applications included spam detection and sentiment analysis.
4. Deep Learning Era (2015–Present)
Neural networks, RNNs, LSTMs, and Transformers revolutionized NLP.
Models like BERT and GPT achieved state-of-the-art performance.
Conclusion:

NLP has evolved from rule-based systems to advanced deep learning models capable of understanding and generating human-like text.

Question 4: What is text normalization and why is it essential in text processing tasks?
Answer:

Text normalization is the process of converting text into a standard and consistent format before analysis.

Common normalization techniques:
Converting text to lowercase
Removing punctuation
Removing stopwords
Expanding contractions
Lemmatization or stemming
Example:

Original Text:

Running, RUNNING, runs!

Normalized Text:

run run run
Importance:
Reduces data inconsistencies.
Improves model accuracy.
Simplifies text processing.
Reduces vocabulary size.
Conclusion:

Text normalization is a fundamental preprocessing step that ensures clean and consistent text data for NLP applications.

Question 5: Compare and contrast stemming and lemmatization with suitable examples.
Answer:
Stemming:

Stemming removes prefixes or suffixes from words to obtain the root form without considering grammar.

Examples:

Running → Run
Studies → Studi
Playing → Play
Lemmatization:

Lemmatization converts words into their meaningful dictionary form (lemma) using linguistic analysis.

Examples:

Running → Run
Studies → Study
Better → Good
Difference Table:
Feature	Stemming	Lemmatization
Method	Rule-based cutting	Dictionary-based
Accuracy	Lower	Higher
Speed	Faster	Slower
Output	May not be meaningful	Meaningful words
Conclusion:

Stemming is faster but less accurate, while lemmatization produces linguistically correct words and is generally preferred for modern NLP tasks.

Question 6: Write a Python program that uses regular expressions (regex) to extract all email addresses.
Answer:
Python Code:
import re

text = """
Hello team, please contact us at support@xyz.com for technical issues,
or reach out to our HR at hr@xyz.com.
You can also connect with John at john.doe@xyz.org and jenny via
jenny_clarke126@mail.co.us.

"""

for email in emails:
    print(email)

Question 7: Perform string tokenization and frequency distribution using Python and NLTK.
Answer:
Python Code:
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

nltk.download('punkt')

text = """
Natural Language Processing (NLP) is a fascinating field that combines
linguistics, computer science, and artificial intelligence.
It enables machines to understand, interpret, and generate human language.
Applications of NLP include chatbots, sentiment analysis,
and machine translation.
As technology advances, the role of NLP in modern solutions
is becoming increasingly critical.
"""

tokens = word_tokenize(text)

fdist = FreqDist(tokens)

print("Tokens:")
print(tokens)

print("\nFrequency Distribution:")
print(fdist.most_common(10))

'fascinating', 'field', ...]

Frequency Distribution:
[('NLP', 3),
 ('and', 3),
 ('is', 2),
 ('of', 2),
 ('Natural', 1),
 ('Language', 1),
 ('Processing', 1)]

Question 9: Train a simple Word2Vec model using Gensim.
Answer:
Python Code:
from gensim.models import Word2Vec

dataset = [
    "Natural language processing enables computers to understand human language",
    "Word embeddings are a type of word representation that allows words with similar meaning to have similar representation",
    "Word2Vec is a popular word embedding technique used in many NLP applications",
    "Text preprocessing is a critical step before training word embeddings",
    "Tokenization and normalization help clean raw text for modeling"

from textblob import TextBlob

reviews = [
    "The app is very easy to use.",
    "Customer support is terrible.",
    "Loan approval process was quick and smooth."
]

    sentiment = TextBlob(review).sentiment.polarity

## Question 9: Train a simple Word2Vec model using Gensim.

### Python Code:

```python
!pip install gensim
from gensim.models import Word2Vec

dataset = [
    "Natural language processing enables computers to understand human language",
    "Word embeddings are a type of word representation that allows words with similar meaning to have similar representation",
    "Word2Vec is a popular word embedding technique used in many NLP applications",
    "Text preprocessing is a critical step before training word embeddings",
    "Tokenization and normalization help clean raw text for modeling"
]

processed_data = [sentence.lower().split() for sentence in dataset]

model = Word2Vec(
    sentences=processed_data,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)

print("Vocabulary:")
print(list(model.wv.key_to_index.keys()))

print("\nVector for word 'word':")
print(model.wv['word'][:10])
```

## Question 10: Customer Feedback Analysis Using NLP

As a data scientist at a fintech startup, I would follow the following steps:

### Step 1: Data Collection
Gather customer reviews from websites, mobile apps, surveys, and support tickets.

### Step 2: Data Cleaning
*   Remove duplicates.
*   Remove special characters.
*   Convert text to lowercase.
*   Handle missing values.

### Step 3: Text Preprocessing
*   Tokenization

*   Text normalization

### Step 4: Feature Extraction
*   Bag of Words (BoW)

### Step 5: Sentiment Analysis
Classify reviews as Positive, Negative, or Neutral.

### Step 6: Topic Modeling
Identify common themes such as:
*   Customer Support

### Sample Python Code:

```python
from textblob import TextBlob

reviews = [
    "The app is very easy to use.",
    "Customer support is terrible.",
    "Loan approval process was quick and smooth."
]

for review in reviews:
    sentiment = TextBlob(review).sentiment.polarity

    if sentiment > 0:
        label = "Positive"
    elif sentiment < 0:
        label = "Negative"
    else:
        label = "Neutral"

    print(review, "->", label)
```
